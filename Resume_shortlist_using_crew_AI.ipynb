```python
from crewai import Agent,Task,Crew
from langchain_community.chat_models import ChatLiteLLM
import os
```

```python
os.environ["GOOGLE_API_KEY"] = "AIzaSyDWr5lAx5AlEcM096H7DoGvJmGclM9rIVQ"
```

```python
llm = ChatLiteLLM(
    model="gemini/gemini-1.5-flash",
    temperature=0,
    api_key = os.environ["GOOGLE_API_KEY"]
)
```

```python
resume_text = """ Name: A R Arulsha
  Skills: Languages : Python, R(basics), HTML, CSS
          Frameworks : Pandas, NumPy, TensorFlow, Kera’s, Scikit-Learn, Matplotlib and Seaborn, MLib, Apace Spark
          Databases : MySQL
          Data Visualization Tools : Tableau , PowerBi , Qlik Sense
          Platforms : Google Collab, MongoDB, Visual Studio Code, Qlik Sense, Apache Spark, Kaggle, Docker, GitHub
          Soft Skills : Excellent Communication, Teamwork, Leadership
  Experience: 3 years
  Degreee : B.tech Artificial Intelligence And Data Science"""
```

```python
#Agent 1: Evaluator
evaluator = Agent(
    role = "Resume Evaluator",
    goal = "Evaluate resumes for AI/Ml job fit",
    backstory = "Expert in talent assessment and AI hiring standards.",
    verbose = True,
    llm=llm
)
```

```python
#Agent 2: Reply Writer
writer =Agent(
    role = "Reply Writer",
    goal = "Write clear replies to candidates",
    backstory = "An HR pro who crafts polite offers/reject messages.",
    verbose = True,
    llm=llm
)
```

```python
#Task 1 : Evaluation
evaluation_task = Task(
    description = f"""check if the following resume is fit for an AI/ML Engineer role. Respond with YES or NO and a short reason:\n\n{resume_text}""",
    agent = evaluator,
    expected_output = "YES or NO with short reason with one reply"
)
```

```python
# TASK 2 : Write reply
reply_task = Task(
    description = """Using the evaluation result, Write:
    -An acceptance message if YES
    -A polite rejection message if NO""",
    agent = writer,
    context =[evaluation_task],
    expected_output = "A professional message to send to the candidate I just need one reply"
)
```

```python
crew = Crew(
    agents =[evaluator,writer],
    tasks = [evaluation_task,reply_task],
    verbose  = True
)
```

```python
result = crew.kickoff()
print("\n Final Message to caandidate:\n")
print(result)
```
